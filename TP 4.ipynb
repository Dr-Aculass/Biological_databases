```python
import copy
import math

# List of sequence for which we want to compute our matrix.
seq_list = ["FPTADAGGRS", "FVTADALGRS", "FPTPDAGLRN","FVTAEAGLRQ","FPTAEAGGRS"]

amino_acids_keys=list('ACDEFGHIKLMNPQRSTVWY')
amino_acids = dict.fromkeys(amino_acids_keys, 0)
amino_acids_freq = dict.fromkeys(amino_acids_keys, 0)
amino_acids_matrix = dict()
amino_acids_count = 0

# Compute frequency and occurrences of aminoacids.
for seq in seq_list:
    for char in seq:
       if char in amino_acids:
           amino_acids_count += 1
           amino_acids[char] += 1
            
for key in amino_acids:
    amino_acids_freq[key] = amino_acids[key]/amino_acids_count

# Matrix of the cAB value.
for i in range(len(amino_acids_keys)):
    for j in range (len(amino_acids_keys)):
        amino_acids_matrix[amino_acids_keys[i]+amino_acids_keys[j]] = 0
        
# Initiate two dictionnary we will need later. 
amino_acids_matrix_e = copy.deepcopy(amino_acids_matrix)
amino_acids_matrix_score = copy.deepcopy(amino_acids_matrix)

# Compute the cAB value.
for k in range(len(max(seq_list))):
    for i in range(len(seq_list)):
        for j in range(i+1, len(seq_list)):
            amino_acids_matrix[seq_list[i][k]+seq_list[j][k]] = amino_acids_matrix[seq_list[i][k]+seq_list[j][k]] + 1
            amino_acids_matrix[seq_list[j][k]+seq_list[i][k]] = amino_acids_matrix[seq_list[j][k]+seq_list[i][k]] + 1

# As our previous computation allowed some value (Diagonal line) to be counted 2 times, divide those value by 2. 
for ai in amino_acids_keys:
    amino_acids_matrix[ai+ai] = int(amino_acids_matrix[ai+ai]/2)
    

T = amino_acids_count *2

# Initate the matrix to compute qAB value.
amino_acids_matrix_q = copy.deepcopy(amino_acids_matrix)

# Compute qAB values.
for ai in amino_acids_keys:
    for aj in amino_acids_keys:
        amino_acids_matrix_q[ai+aj] = amino_acids_matrix[ai+aj]/T

#Initiate a new dictionary from our amino acids.        
p = dict.fromkeys(amino_acids_keys, 0)

# Divide the values obtained previously by 100 to 
for ai in amino_acids_keys:
    for aj in amino_acids_keys:
        if ai == aj:
            p[ai] += amino_acids_matrix[ai+aj]/100
        else:
            p[ai] += amino_acids_matrix[ai+aj]/200


# Compute de eAB values.             
for ai in amino_acids_keys:
    for aj in amino_acids_keys:
        if ai == aj:
            amino_acids_matrix_e[ai+aj] = p[ai]*p[ai]
        else:
            amino_acids_matrix_e[ai+aj] = 2*(p[ai]*p[aj])

# Compute our final matrix.            
ex=0
for k in range(len(max(seq_list))):
    for i in range(len(seq_list)):
        for j in range(i+1, len(seq_list)):
            ex = amino_acids_matrix_q[seq_list[i][k]+seq_list[j][k]]/amino_acids_matrix_e[seq_list[i][k]+seq_list[j][k]]
            amino_acids_matrix_score[seq_list[i][k]+seq_list[j][k]]=round(2*math.log(ex,2))


print(amino_acids_matrix)
print(amino_acids_matrix_q)
print(p)
print(amino_acids_matrix_e)
print (amino_acids_matrix_score)
```

```python
import pandas as pd

# file should be in the same directory.
def create_PAM(file,xpo):
    with open(file, 'rt') as f:
        csvFile = pd.read_csv(f, sep=',', header=None)
        x = csvFile
        for i in range(1,len(x[0])):
            for j in range(1,len(x[0])):
                if type(x[i][j]) == str:
                    x[i][j]= int(float(x[i][j]))
                    x[i][j] = (x[i][j]/10000)**xpo
                    
        print(x)
        
create_PAM("TP4pam1.csv", xpo=150)
    
    
```
